## Loading & Preprocessing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



df=pd.read_csv("https://raw.githubusercontent.com/RICKY-UPB/data/main/Assignment-2_Data.csv")
df.head()
```

```python
#Cek duplikasi data
df.duplicated().sum()
```

```python
#Jika ada data dupilkasi,maka dihapus
df.drop_duplicates(inplace=True)
```

```python
#Hapus kolom yang unik nilainya
#df.drop(columns=['pdays'], inplace=True)
```

```python
#Cek Data yang missing value/kosong/null/NA/NaN

print('Jumlah baris dan kolom : ', df.shape)
df.isna().sum()

```

```python
#Cek baris yang nilai Age, job nya kosong/NaN
df[df.age.isna()]
```

```python
#Isi data kosong age dengan nilai median
df.age.fillna(df.age.median(), inplace=True)
```

```python
#Cek nilai yang sering muncul pada kolom job
df.job.value_counts()
```

```python
#Isi data kosong dengan nilai 'blue collar'(sering muncul) di  data job
df.job.fillna('blue-collar', inplace=True)

#cek kembali data missing Value
print('Jumlah baris dan kolom : ', df.shape)
df.isna().sum()
```

```python
#Cek baris yang nilai balance nya kosong/NaN
df[df.balance.isna()]
```

```python
#Jika missing value hapus baris
df.dropna(inplace=True)

#cek kembali data missing Value
print('Jumlah baris dan kolom : ', df.shape)
df.isna().sum()
```

# Chart Distribution EDA (Exploratory Data Analysis)

### Import Library

```python
#import library seaborn dan matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

## KATEGORIKAL TYPE

### Mengecek distribusi data menggunakan sns counplot pada data kategorikal

```python
col_of_data = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']



for i in col_of_data:
    plt.figure(figsize=(7,4))
    plt.title(i)
    sns.countplot(data=df, y=i, palette='Set2')
    plt.show()
```

### Level pada data Kategorikal

```python
# Mengecek level pada kolom kategorikal

cols_cat = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']

for col in cols_cat:
    print(f'Column {col}: {df[col].nunique()} sublevels')
```

## Numeric data type chart


### Mengecek distribusi data menggunakan Boxplot (Mengecek nilai nilai yang berada diluar rentang)

```python
#Membuat Boxplot distribusi data Numerical keseluruhan
cols_num = ['Id', 'age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

#fig, ax = plt.subplots(nrows=8, ncols=1, figsize=(8,30))


fig, axs = plt.subplots(4,2, figsize=(16, 8))


for col, ax in zip(cols_num, axs.flatten()):
    ax.set_title(col)
    sns.boxplot(x=col, data=df, ax=ax)

plt.tight_layout()
plt.show()
```

### Mengeliminasi Data yang tidak relevan (berada diluar rentang)



```python
# terdapat nilai data yang diluar rentang yaitu kolom age dan previous sehingga kita eliminasi

# eliminasi data Age yang berumur lebih dari 100 dan kurang dari 18
print(f' Jumlah dataset sebelum eliminasi baris age : {df.shape}')
df = df[df['age']<=100]
df = df[df['age']>=18]
print(f'Jumlah dataset sesudah eliminasi baris age : {df.shape}')
```
